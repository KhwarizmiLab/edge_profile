## Arch Prediction

```python
from data_engineering import all_data
from architecture_prediction import get_arch_pred_model
from sklearn.metrics import classification_report
```

```python
results = {
    "Models": [],
    "Train Accuracy": [],
    "Test Accuracy": []
}
```

```python
from pathlib import Path
data = all_data(Path("/content/drive/MyDrive/EZClone/edge_profile/profiles/tesla_t4/colab_zero_pretrained_all_noexe/"))
data.head()
```

### Random Forest

```python
arch_pred_model = get_arch_pred_model("rf", df=data)
results["Models"].append("Random Forest")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```

### KNN

```python
arch_pred_model = get_arch_pred_model("knn", df=data)
results["Models"].append("K Nearest Neighbors")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```

### Neural Network

```python
arch_pred_model = get_arch_pred_model("nn", df=data)
results["Models"].append("Neural Network")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```

```python
arch_pred_model = get_arch_pred_model("lr", df=data)
results["Models"].append("Logistic Regression")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```

### Nearest Centroid

```python
arch_pred_model = get_arch_pred_model("centroid", df=data)
results["Models"].append("Nearest Centroid")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```

### Naive Bayes

```python
arch_pred_model = get_arch_pred_model("nb", df=data)
results["Models"].append("Naive Bayes")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```

### AdaBoost

```python
arch_pred_model = get_arch_pred_model("ab", df=data)
results["Models"].append("AdaBoost")
results['Test Accuracy'].append(arch_pred_model.evaluateTest())
results['Train Accuracy'].append(arch_pred_model.evaluateTrain())
```
